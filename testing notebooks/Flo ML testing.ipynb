### From the paper: 

Smoothing Splines
The smoothing splines regression was trained using the function “train” from the library “caret” of R language, using method equal to “bstSM”. A k-fold cross-validation system was used, with k = 10. The predictors used to construct the model were t, Tinj, Tgrw, M/C, and [QD]. Other predictors were not used, because the function does not work properly in the presence of NA values nor correlated variables. There was no further treatment such as scaling or normalization.
